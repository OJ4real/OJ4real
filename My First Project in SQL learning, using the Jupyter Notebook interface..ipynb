# The 'facts' Table.

So from the output derived earlier, the 'facts' table happens to be the only table stored within the Factbook.db. This table contains several coulmns and rows. To understand what it looks like, the first 5 rows will be retreived.

```python
%%sql #it is very important to always start your querry with '%%sql' code as the first line of code.
SELECT *
    FROM facts
    LIMIT 5;
```

# The Minimum & Maximum population, as well as the Minimum and Maximum population growth within the 'facts' table.

```python
%%sql
SELECT MIN(population) 'Lowest population', MAX(population)'Highest population', 
       MIN(population_growth), MAX(population_growth)
From facts;
```

# The countries with the Highest and Lowest population.

```python
%%sql
SELECT name
From facts
where population= (SELECT MIN(population)
             From facts);
```

```python
%%sql
SELECT name
From facts
where population= (SELECT MAX(population)
             From facts);
```

The querry for the name of the country with the Highest population unfortunately produced the name 'world', which doesn't look right. This is so because a value was entered for the name world in the 'facts' table. Nevertheless, this can be corrected with the codes below.

```python
%%sql
SELECT MIN(population), Max(population), MIN(population_growth),Max(population_growth)
    FROM facts
    Where name!= 'World';
```

#  The average population and area of the 'facts' table.

```python
%%sql
SELECT AVG(population), AVG(area)
From facts;
```

# The countries below the average population index.

```python
%%sql
SELECT id,code,name
From facts
Where population > (SELECT AVG(population) and area < (AVG(area))
                   from facts);
```
